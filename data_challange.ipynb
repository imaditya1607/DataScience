```python
import pandas as pd #importing pandas 
```

```python
raw_data = pd.read_csv('data_cleaning_challenge.csv') #importing my csv file
```

```python
raw_data.head(20)
```

```python
data_set_1 = raw_data.drop(columns=['Unnamed: 9','Unnamed: 10']) # dropping two column which is having non values 
```

```python
# there is subset of data in this data frame 
#    so will try to make groups in data 
data_set_2=data_set_1[data_set_1['Row Type'].notna()]
```

```python
column_values = []
counter = 0
for values in data_set_2['Row Type']:
  if "first name" in values :
    counter+=1
  column_values.append(counter)
```

```python
df=pd.DataFrame(column_values)
df.tail(1)
```

```python
data_set_2.insert(0,"Iteration",value = column_values)
```

```python
data_set_3 = data_set_2[data_set_2['Row Type'] != "Row Type"]
```

```python
data_set_4 = data_set_3[data_set_3['Row Type'].str.contains('first name')]
```

```python
data_set_4.drop(columns=['Speed1','Speed2','Electricity','Effort','Weight','Torque'], inplace = True)
```

```python
data_set_4.rename(columns={'Row Type':'First Name',	'Iter Number': 'Last Name',	'Power1' : 'Date'}, inplace = True)
```

```python
data_set_4['First Name']=data_set_4['First Name'].str[12:]
data_set_4['Last Name']=data_set_4['Last Name'].str[11:]
data_set_4['Date']=data_set_4['Date'].str[5:]
```

```python
data_set_5=data_set_3[~data_set_3['Row Type'].str.contains('first name')]
```

```python
data_set_final = pd.merge(left=data_set_4 , right=data_set_5 , how="inner" , on="Iteration")
```

```python
data_set_final.drop(columns=["Iteration"], inplace=True )
data_set_final
```
